```python
import matplotlib.pyplot as plt
from collections import Counter
import glob

import numpy as np
```

```python
!cargo run --bin histogram-data data/bb3_release/RV30/BBS30002.xml test.out 10011001
```

```python
counter_neg = Counter()
counter_part = Counter()
counter_pos = Counter()

with open('test.out') as f:
    lines = f.readlines()
    for line in lines:
        score, label = line.split(';')
        score = int(score)
        label = label.strip()
        counter_neg[score] = 0
        counter_part[score] = 0
        counter_pos[score] = 0
        
    
    for line in lines:
        score, label = line.split(';')
        score = int(score)
        label = label.strip()
        if label == '-1':
            counter_neg[score] += 1
        elif label == '0':
            counter_part[score] += 1
        else:
            counter_pos[score] += 1

neg_scores, neg_occurences = list(zip(*counter_neg.items()))
part_scores, part_occurences = list(zip(*counter_part.items()))
pos_scores, pos_occurences = list(zip(*counter_pos.items()))

neg_occurences = np.array(neg_occurences)
part_occurences = np.array(part_occurences)
pos_occurences = np.array(pos_occurences)
```

```python
for path in glob.glob('data/bb3_release/RV30/BBS*.xml'):
    !cargo run --bin histogram-data {path} test.out 0001110101
    
    counter_neg = Counter()
    counter_part = Counter()
    counter_pos = Counter()

    with open('test.out') as f:
        lines = f.readlines()
        for line in lines:
            score, label = line.split(';')
            score = int(score)
            label = label.strip()
            counter_neg[score] = 0
            counter_part[score] = 0
            counter_pos[score] = 0


        for line in lines:
            score, label = line.split(';')
            score = int(score)
            label = label.strip()
            if label == '-1':
                counter_neg[score] += 1
            elif label == '0':
                counter_part[score] += 1
            else:
                counter_pos[score] += 1

    neg_scores, neg_occurences = list(zip(*counter_neg.items()))
    part_scores, part_occurences = list(zip(*counter_part.items()))
    pos_scores, pos_occurences = list(zip(*counter_pos.items()))

    neg_occurences = np.array(neg_occurences)
    part_occurences = np.array(part_occurences)
    pos_occurences = np.array(pos_occurences)
    
    plt.figure()
    plt.bar(pos_scores, pos_occurences, color='green')
    plt.bar(part_scores, part_occurences, bottom=pos_occurences, color='orange')
    plt.bar(neg_scores, neg_occurences, bottom=pos_occurences+part_occurences, color='red')
```
